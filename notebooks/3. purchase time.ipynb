creating a dataframe with all the information about the loyal customer

```python
loyal_info=pd.DataFrame(columns=['order_id','user_id','created_at','discount_cost','discount_type','basket',
                                 'month','day','year','cohort'])
```

```python
loyal_info=pd.concat([loyal_info,data[data['user_id']==1984]])
```

```python
for i in range(1250):
    user=df['user_id'][i]
    loyal_info=pd.concat([loyal_info,data[data['user_id']==user]])
```

```python
loyal_info=loyal_info.reset_index()
```

```python
loyal_info.drop('index',axis=1,inplace=True)
```

```python
loyal_info[loyal_info['user_id']==3276]
```

```python
loyal_info['purchase_day']=(loyal_info['year']-2020)*365+loyal_info['month']*30+loyal_info['day']
```

```python
loyal_info[loyal_info['user_id']==3276].sort_values(by=['year','month','day'])
```

```python
second_purchase=[]
third_purchase=[]
fourth_purchase=[]
fifth_purchase=[]
for i in range(1250):
    user=df['user_id'][i]
    first=loyal_info[loyal_info['user_id']==user].sort_values(['year','month','day'])['purchase_day'].iloc[0]
    second=loyal_info[loyal_info['user_id']==user].sort_values(['year','month','day'])['purchase_day'].iloc[1]
    third=loyal_info[loyal_info['user_id']==user].sort_values(['year','month','day'])['purchase_day'].iloc[2]
    fourth=loyal_info[loyal_info['user_id']==user].sort_values(['year','month','day'])['purchase_day'].iloc[3]
    fifth=loyal_info[loyal_info['user_id']==user].sort_values(['year','month','day'])['purchase_day'].iloc[4]
    first_second=second-first
    second_third=third-second
    third_fourth=fourth-third
    fourth_fifth=fifth-fourth
    second_purchase.append(first_second)
    third_purchase.append(second_third)
    fourth_purchase.append(third_fourth)
    fifth_purchase.append(fourth_fifth)
```

```python
f, ((ax1, ax2),(ax3,ax4)) = plt.subplots(2, 2, sharey=True,squeeze=False)
ax1.hist(second_purchase,bins=30);
ax2.hist(third_purchase,bins=30);
ax3.hist(fourth_purchase,bins=30);
ax4.hist(fifth_purchase,bins=30);

plt.subplots_adjust(left=0.1,bottom=0.1,right=0.9,top=0.9,wspace=0.4,hspace=0.4);

#######
ax1.set_title('first and second purchase')
ax2.set_title('second and third purchase')
ax3.set_title('third and fourth purchase')
ax4.set_title('fourth and fifth purchase')
```
